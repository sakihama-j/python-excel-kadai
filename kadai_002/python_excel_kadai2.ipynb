```python
import pandas as pd

# データフレームの作成
df = pd.DataFrame({
    '日付': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'], # 課題の日付を入力
    '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],      # 課題の社員名を入力
    '売上': [100, 200, 150, 300, 250], #Pythonおよびpandasでは、クォーテーションで囲むと文字列として扱われます。文字列のままだと、その後の計算（mean()での平均算出）や、関数内での比較（sales >= ...）でエラーが発生します。
    '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店'],
})


# 平均を求めて新しい列を作成
df['平均売上'] = df['売上'].mean()

# 平均値を計算しておく（条件判定に使うため）
aveage_sales = df['売上'].mean()

def performance(sales):
  result = ''
  if sales >= aveage_sales + 50: # 平均売上+50以上
    result = 'A'
  elif sales >= aveage_sales:
    result = 'B'
  else:
    result = 'C'
  return result

# 「業績ランク」列を作成し、関数「performance」を適用
df['業績ランク'] = df['売上'].apply(performance)

#Excelファイルを作成
writer = pd.ExcelWriter('業績.xlsx')

# DataFrameオブジェクトをExcelファイルに書き込む（インデックスは含めない指定）
df.to_excel(writer, sheet_name='Sheet1', index=False)

# Excelファイルを閉じる
writer.close()
```
